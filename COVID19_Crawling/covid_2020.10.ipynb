```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import pandas as pd
import pandas
import numpy
from bs4 import BeautifulSoup
import lxml

import time

```

```python
path = 'C:\\Users\\chromedriver.exe' #옵션 + 커맨드 + c
driver = webdriver.Chrome(path)
driver.implicitly_wait(3)

start_url = 'https://search.naver.com/search.naver?where=blog&query=%EC%BD%94%EB%A1%9C%EB%82%98%20%EB%B0%B1%EC%8B%A0&sm=tab_opt&nso=so%3Ar%2Cp%3Afrom20201001to20201031'
driver.get(start_url)
```

```python
SCROLL_PAUSE_TIME = 0.5

body = driver.find_element_by_tag_name('body')

while True:
    last_height = driver.execute_script('return document.documentElement.scrollHeight')
    #현재 화면의 길이를 리턴받아서 last_height
    
    for i in range(10):
        body.send_keys(Keys.END) #body 본문에 END를 넣어주는거 (스크롤 내려줌)
        
        time.sleep(SCROLL_PAUSE_TIME) #데이터로딩시간을 주기위해서 스크롤 후 휴식
        
    new_height = driver.execute_script('return document.documentElement.scrollHeight')
    if new_height ==last_height:
        break;
```

```python
page = driver.page_source

soup = BeautifulSoup(page, 'html.parser')

driver.close()
```

```python
len(soup.text)
```

```python
name = soup.find_all('a', attrs={'class':'sub_txt sub_name'})
name[0].get_text()

name_list = []

for i in name:
    name_list.append(i.get_text())
print(name_list)
```

```python
len(name_list)
```

```python
title = soup.find_all('a', attrs={'class':'api_txt_lines total_tit'})
title[0].get_text()

title_list = []

for i in title:
    title_list.append(i.get_text())
print(title_list)
```

```python
review = soup.find_all('div', attrs={'class':'api_txt_lines dsc_txt'})
review[0].get_text()

review_list = []

for i in review:
    review_list.append(i.get_text())
print(review_list)
```

```python
date = soup.find_all('span', attrs={'class':'sub_time sub_txt'})
date[0].get_text()

date_list = []

for i in date:
    date_list.append(i.get_text())
print(date_list)
```

```python
covid_2020_10_df = pd.DataFrame({'날짜':date_list, '제목':title_list, '내용':review_list, '작성자':name_list})

covid_2020_10_df
```
